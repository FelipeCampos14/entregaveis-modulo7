```python
import pandas as pd
import numpy as np
import statistics as st
from math import sqrt
```

```python
# Cria o conjuto de dados, o Dataframe
dici = {'Id':['P1','P2','P3','P4','P5','Novo Paciente'],'Sexo':['F','F','M','F','M','F'], 'Cirurgia?':['N','N','S','N','N','N'],'TSH':[6.8,7.3,8.8,6.9,5.4,7.0],'TT4':[156.2,152.9,148.4,132.7,150.9,150],'TI':['N','S','N','N','N','S'],'Dispositivo':['Neg','Neg','HipoPrim','HipoComp','Neg',np.nan]}
df = pd.DataFrame(data=dici)
df
```

# Transformando dados qualitativos em quantitativos

```python
# Replace nos dados, transformandos os dados em bin√°rios
df['Sexo'] = df['Sexo'].replace({'F':0,
                            'M':1})
df['Cirurgia?'] = df['Cirurgia?'].replace({'N':0,
                                            'S':1})
df['TI'] = df['TI'].replace({'N':0,
                            'S':1})
df['Dispositivo'] = df['Dispositivo'].replace({'Neg':0,
                                                'HipoPrim':1,
                                                'HipoComp':2})
df
```

# Normalizando os dados

```python
# Normaliza√ß√£o da coluna: TSH
lista = list()
dici = dict()
ordem = df['TSH'].sort_values().unique()
for i in df['TSH']:
    x = (int(i) - ordem[0])/(ordem[-1]-ordem[0])
    lista.append(x)
df['TSH'] = lista
df
```

```python
# Normaliza√ß√£o da coluna TT4
lista = list()
dici = dict()
ordem = df['TT4'].sort_values().unique()
for i in df['TT4']:
    x = (int(i) - ordem[0])/(ordem[-1]-ordem[0])
    lista.append(x)
df['TT4'] = lista
df
```

# Calculando dist√¢ncia Euclidiana do Novo paciente

```python
lista = list()
dici = dict()
for a,b,c,d,e in zip(df['TSH'],df['TT4'],df['Sexo'],df['Cirurgia?'],df['TI']):
    x = sqrt(((a-df['TSH'].loc[df['Id']=='Novo Paciente'])**2)+((b-df['TT4'].loc[df['Id']=='Novo Paciente'])**2)+((c-df['Sexo'].loc[df['Id']=='Novo Paciente'])**2)+((d-df['Cirurgia?'].loc[df['Id']=='Novo Paciente'])**2)+((e-df['TI'].loc[df['Id']=='Novo Paciente'])**2))
    print(x)
df
```
